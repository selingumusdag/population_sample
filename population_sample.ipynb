```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(8,5))

pop1 = np.random.normal(100, 20, 1000)
pop2 = np.random.normal(60,10, 1000) 

#sample1 = np.random.choice(pop1, 50, replace=True)
#sample2 = np.random.choice(pop2, 1000, replace=True)
```

```python
sample1_mean=[]
sample2_mean=[]

for x in range(1,200):
    sample1 = np.random.choice(pop1, x, replace=True)
    sample1_mean.append(np.mean(sample1))


for x in range(1,100):
    sample2 = np.random.choice(pop2, x, replace=True)
    sample2_mean.append(np.mean(sample2))

```

```python
title_font={'family':'arial','color':'red','weight':'bold','size':15}
axis_font={'family':'arial','color':'black','weight':'bold','size':13}

plt.figure(figsize=(8,4), dpi = 100)
plt.subplot(211)
plt.title('Sample1_Mean Data Histogram',fontdict=title_font)
plt.hist(sample1_mean,bins=50)

plt.figure(figsize=(8,4), dpi = 100)
plt.subplot(212)
plt.title('Sample2_Mean Data Histogram',fontdict=title_font)
plt.hist(sample2_mean,bins=50)
plt.show()
```

```python
print("Mean of Sample1 : ", sample1.mean())
print("Mean of Sample2 : ", sample2.mean())
print("Standard Deviation of Sample1 : ", sample1.std())
print("Standard Deviation of Sample2 : ", sample2.std())
```
